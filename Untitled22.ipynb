```python
from fastai.vision.all import *
from fastai.callback.fp16 import *
import torch
import os
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/histopathological image dataset for ET'
```

```python
print(os.listdir(path))
```

```python
dblock = DataBlock(blocks = (ImageBlock, CategoryBlock),
                   get_items = get_image_files,
                   get_y = parent_label,
                   splitter = RandomSplitter(seed=42, valid_pct=0.2),
                   item_tfms = Resize(224),
                   batch_tfms = aug_transforms(mult=2))
dls = dblock.dataloaders(path, bs=64)
```

```python
dls.show_batch(nrows=2, ncols=2)
```

```python
average = 'macro'
metrics = [accuracy, Precision(average=average), Recall(average=average), F1Score(average=average)]
```

```python
learn50 = vision_learner(dls, resnet50, metrics=metrics, cbs=[GradientAccumulation(64)]).to_fp16()
learn50.fine_tune(10, 1e-3)
```

```python
interp = ClassificationInterpretation.from_learner(learn50)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(9, nrows=3)
```
